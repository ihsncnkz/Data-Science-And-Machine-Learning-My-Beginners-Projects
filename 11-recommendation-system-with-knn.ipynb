In this chapter, I will prapere the dataset for the Recommemdation System. I will apply two ways. The first way is the way I researched. The second way is the way I learned from Udemy courses. I will prepare the dataset for the this ways.

In this code below, I will merge the dataset that ı will use.

```python
data.drop(["genres","timestamp"], axis = 1,inplace = True)
data
```

I make a dataset for the first way.

```python
data_pivot = data.pivot_table(index='userId',columns='title',values='rating')
data_pivot
```

I make a dataset for the second way.

```python
data_pivot2 = pd.pivot(index = "movieId", columns = "userId", data = rating, values = "rating")
data_pivot2
```

```python
Movie_voted = pd.DataFrame(rating.groupby("movieId")["rating"].agg("count"))
Movie_voted.reset_index(level = 0, inplace = True)
Movie_voted
```

```python
User_Voted = pd.DataFrame(rating.groupby("userId")["rating"].agg("count"))
User_Voted.reset_index(level = 0, inplace = True)
User_Voted
```

```python
data_pivot2.fillna(0, inplace = True)
data_pivot2
```

This is the final version of the dataset I will use.

```python
dataLast = data_pivot2.loc[Movie_voted[Movie_voted["rating"] > 10]["movieId"],:]
dataLast = dataLast.loc[:, User_Voted[User_Voted["rating"] > 60]["userId"]]
dataLast
```

# Recommendation Systems <a id = "3"></a>

**What is The Recommendation System?**<br>

Recommender systems are the systems that are designed to recommend things to the user based on many different factors. 

## First The Way

I make a matrix with the last dataset values in the code below.

```python
from scipy.sparse import csr_matrix
csr_data = csr_matrix(dataLast.values)
dataLast.reset_index(inplace=True)
```

```python
csr_data
```

```python
from sklearn.neighbors import NearestNeighbors
knn = NearestNeighbors(metric='cosine', algorithm='brute', n_neighbors=20, n_jobs=-1)
knn.fit(csr_data)
```

It is a Recommendation System with the function below.

```python
def movie_recommendation(movie_name):
    
    n_movies_to_reccomend = 10
    movie_list = movie[movie['title'].str.contains(movie_name)] # Fint movie name
    
    if len(movie_list):
        
        movie_idx= movie_list.iloc[0]['movieId']
        movie_idx = dataLast[dataLast['movieId'] == movie_idx].index[0]
        
        distances , indices = knn.kneighbors(csr_data[movie_idx],n_neighbors=n_movies_to_reccomend+1)    
        rec_movie_indices = sorted(list(zip(indices.squeeze().tolist(),distances.squeeze().tolist())),key=lambda x: x[1])[:0:-1]
        recommend_frame = []
        
        for val in rec_movie_indices:
            movie_idx = dataLast.iloc[val[0]]['movieId']
            idx = movie[movie['movieId'] == movie_idx].index
            recommend_frame.append({'Title':movie.iloc[idx]['title'].values[0],'Distance':val[1]})
        df = pd.DataFrame(recommend_frame,index=range(1,n_movies_to_reccomend+1))
        return df
    
    else:
        return "No movies found. Please check your input"
```

```python
# result of Recommendation System 
movie_recommendation('Gladiator')
```
